```python
def fix_platforms(x):
    windows = 0
    mac = 0
    linux = 0
    if type(x) == str:
        data = eval(x)
        if data['windows'] == True:
            windows = 1
        if data['mac'] == True:
            mac = 1
        if data['linux'] == True:
            linux = 1
    return windows, mac, linux
        
df['platforms'].unique() #One hot encode these 
```

```python
res = df['platforms'].apply(fix_platforms)
```

```python
windows = []
mac = []
linux = []
for tup in res:
    windows.append(tup[0])
    mac.append(tup[1])
    linux.append(tup[2])
df['windows'] = windows
df['mac'] = mac
df['linux'] = linux
df = df.drop('platforms', axis = 1)
df.head()
```
